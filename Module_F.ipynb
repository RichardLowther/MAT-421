20.2 Finite Difference Approximating Derivatives

```python
# There are three finite difference methods: forward, backward, and central. At a given point at x = a:
# forward difference takes the slope between x = a and x = a + h, where h is a step size
# backward difference takes the slope between x = a and x = a - h
# central difference takes the slope between x = a - h and x = a + h

import matplotlib.pyplot as plt
# step size
h = 0.1
# define grid
x = np.arange(0, 2*np.pi, h)
# compute function
y = np.cos(x)

# compute vector of forward differences
forward_diff = np.diff(y)/h
# compute corresponding grid
x_diff = x[:-1:]
# compute exact solution
exact_solution = -np.sin(x_diff)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
         label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution, \
         label = 'Exact solution')
plt.legend()
plt.show()

# Compute max error between
# numerical derivative and exact solution
max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

20.4 Numerical Differentiation with Noise

```python
# Sometimes data can be contaminated with noise, meaning its value is off by a small amount from what it would be if it were computed from a pure mathematical function
# Example of an equation with minimal noise

%matplotlib inline
x = np.arange(0.01, 2*np.pi, 0.01)
# compute function
omega = 100
epsilon = 0.1

y = np.log(x)
y_noise = y + epsilon*np.cos(omega*x)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'log(x) + noise')
plt.plot(x, y, 'b-', \
         label = 'log(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```

```python
# Example of large amounts of noise

x = np.arange(0, 2*np.pi, 0.01)
# compute function
y = np.cos(x)
y_noise = y + epsilon*omega*np.cos(omega*x)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'Derivative cos(x) + noise')
plt.plot(x, y, 'b-', \
         label = 'Derivative of cos(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```
