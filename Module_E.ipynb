3.2 Continunity and Differentiation

```python
# sin(x^2)/x is a discontinuous function at x = 0. However, we can use Python to find the limit when approaching x = 0
import numpy as np
import matplotlib.pyplot as plt

x = symbols('x')
expr = sin(x**2)/x;
limit_expr = limit(expr, x, 0)
print(format(limit_expr))
```

```python
# Python can also perform derivatives

x = symbols('x')
f = 6*exp(5*x)
derivative_f = f.diff(x)
print(derivative_f)
```

Taylor's Theorem

```python
# The Taylor Series can approximate certain values, in this instance, e^x at x = 0. This technique is especially powerful when there is a point around which we have knowledge about a function for all its derivatives

x = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x))

labels = ['First Order', 'Third Order', 'Fifth Order', 'Seventh Order']

plt.figure(figsize = (10,8))
for n, label in zip(range(4), labels):
    y = y + ((-1)**n * (x)**(2*n+1)) / np.math.factorial(2*n+1)
    plt.plot(x,y, label = label)

plt.plot(x, np.sin(x), 'k', label = 'Analytic')
plt.grid()
plt.title('Taylor Series Approximations of Various Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

Optimization With Gradient Descent

```python
# Gradient Descent is used to find minimum value of a function by repeatedly moving in the direction of steepest descent.

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Define the function
def f(x):
    return 2*x**4 - 6*x**2 + 2*x

# Generate x values for plotting
x_values = np.linspace(-2, 2, 500)
y_values = f(x_values)

# Plot the function
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(False)

# Find global minimum using optimization
result_global = minimize(f, x0=0)  # Start from x=0
x_global_min = result_global.x
y_global_min = result_global.fun

# Plot global minimum
plt.plot(x_global_min, y_global_min, 'bo', label=f'Global Minimizer: ({x_global_min[0]:.2f}, {y_global_min:.2f})')

# Find local minimum using optimization (starting from x=2)
result_local = minimize(f, x0=2)
x_local_min = result_local.x
y_local_min = result_local.fun

# Plot local minimum
plt.plot(x_local_min, y_local_min, 'go', label=f'Local Minimizer: ({x_local_min[0]:.2f}, {y_local_min:.2f})')
plt.legend()
plt.show()
```
