17.2 Linear Interpolation

```python
#Assuming that the series is linear, if we want to find a point that is between our known data

x = [1, 3, 5, 7, 9]
y = [2, 10, 7, 14, 22]

x_interp = 4
y_interp = np.interp(x_interp,x, y)
print(y_interp)
plt.plot(x,y,'-bo',x_interp,y_interp,'ro') #red point is the interpolated point at x=4
plt.title('Data Set with Interpolated Point')
plt.ylim(0,25)
plt.xlabel('x_i')
plt.ylabel('y_i')
plt.show()
```

17.3 Cubic Spline Interpolation

```python
#Assuming that the series is not linear, we can create a smooth curve equation by creating multiple piecewise equations between two points of data
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [1, 3, 5, 7, 9]
y = [2, 10, 7, 14, 22]

func = CubicSpline(x, y, bc_type = 'natural')
x_cubic = np.linspace(0, 9, 100)
y_cubic = func(x_cubic)

plt.plot(x_cubic,y_cubic,'-',x,y,'bo')
plt.title('Data Set with Cubic Spline')
plt.ylim(0,25)
plt.xlabel('x_i')
plt.ylabel('y_i')
plt.show()
```
